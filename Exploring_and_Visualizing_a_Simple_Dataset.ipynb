```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Inspect the dataset
print("Shape of the dataset:", iris.shape)
print("Column names:", iris.columns.tolist())
print("First few rows of the dataset:")
print(iris.head())

# Summary statistics
print("\nSummary information:")
print(iris.info())
print("\nDescriptive statistics:")
print(iris.describe())

# Visualize the dataset

# Scatter plot to show relationships between features
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.grid()
plt.show()

# Histograms to show value distributions
plt.figure(figsize=(12, 8))
iris.hist(bins=20, figsize=(12, 8), layout=(2, 2), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Iris Features')
plt.show()

# Box plots to identify outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris, orient='h')
plt.title('Box Plots of Iris Features')
plt.xlabel('Value')
plt.ylabel('Features')
plt.show()
```
